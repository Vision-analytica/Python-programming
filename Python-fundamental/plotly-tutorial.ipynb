You may encounter some issue while using iplotly. In my case, I got 

1. `ValueError: Mime type rendering requires nbformat>=4.2.0 but it is not installed`
2. `ValueError: Mime type rendering requires nbformat>=4.2.0 but it is not installed`

**First issue is explained below:**  The `ValueError: Mime type rendering requires nbformat>=4.2.0 but it is not installed` error typically occurs when you try to render a Jupyter Notebook file using a package like `nbconvert`, but the `nbformat` package is not installed or is an older version.

`nbformat` is a package that provides a data structure for working with Jupyter Notebook files. It is required by nbconvert to convert or render Jupyter Notebook files to different formats, such as `HTML`, `PDF`, or `Markdown`.

To resolve this error, you can install or upgrade the nbformat package using pip. Here are the steps:

- Open your terminal or command prompt.
- Check if the nbformat package is already installed by running the following command: `pip show nbformat`
   
   If the package is not installed, you will see an error message. Otherwise, you will see information about the package, including the version number.
- If `nbformat` is not installed, or if the version is older than `4.2.0`, you can install or upgrade it using pip: `pip install --upgrade nbformat`.
  
  Once the installation or upgrade is complete, try rendering your Jupyter Notebook file again using nbconvert or any other package that requires nbformat. The error message should no longer appear.

Note that if you are using Jupyter Notebook or JupyterLab, you may not need to install nbformat separately, as it is already included as a dependency of these packages.

**Second issue is explained here:** Here are some troubleshooting steps you can try:

- Make sure you have the latest version of Jupyter Notebook installed. You can check the version by running the following command in your terminal or command prompt:
  `jupyter --version`

  If your version is outdated, you can upgrade to the latest version using pip: `pip install --upgrade jupyter`
- Try creating a new Jupyter Notebook and see if you can render it using `nbconvert`. If you can, the issue may be with the original notebook file you were trying to render.
- Try opening the Jupyter Notebook in a different browser or in a private browsing window to see if the issue is related to your browser cache or cookies.
- If you are running Jupyter Notebook on a remote server, make sure you have the necessary permissions to access and modify files in the directory where the notebook file is located.
- If none of these steps work, you can try uninstalling and reinstalling Jupyter Notebook and its dependencies:

  ```
  pip uninstall jupyter nbconvert nbformat
  pip install jupyter nbconvert nbformat
  ```

  This will remove all Jupyter Notebook-related packages and reinstall them from scratch.

Hopefully one of these steps will resolve the issue and allow you to render your Jupyter Notebook file.

## 3. Bar Charts

````python
df_us = px.data.gapminder().query("country=='United States'")
fig = px.bar(df_us, x='year', y='pop', labels={'year': 'Year', 'pop': 'Population'})
fig.update_layout(title='Population of the United States Over Time')
fig.show()
````

````python
df_tips = px.data.tips()
px.bar(df_tips, x='day', y='tip', color='sex',
       title='Tips by sex on each day',
       labels={'tip':'Tip Amount', 'day': 'Day of the week'})

px.bar(df_tips, x='sex', y='total_bill', color='smoker', barmode='group')
````

````python
df_europe =px.data.gapminder().query("continent=='Europe' and year==2007 and pop>2.e6")
fig=px.bar(df_europe, y='pop', x='country', text='pop', color='country')
fig
````

````python
df_europe =px.data.gapminder().query("continent=='Europe' and year==2007 and pop>2.e6")
fig=px.bar(df_europe, y='pop', x='country', text='pop', color='country')
fig.update_traces(texttemplate='%{text:.2s}',textposition='outside')
fig.update_layout(uniformtext_minsize=8)
fig.update_layout(xaxis_tickangle=-45)

fig
````

## 3. Scatter Plots

````python
df_iris = px.data.iris()
px.scatter(df_iris, 
           x='sepal_width', 
           y='sepal_length', 
           color='species', 
           size ='petal_length',
           hover_data=['petal_width'])
````

````python
df_iris = px.data.iris()
px.scatter(df_iris, 
           x='sepal_width', 
           y='sepal_length', 
           color='species', 
           size ='petal_length',
           hover_data=['petal_width'])
fig=go.Figure()
fig.add_trace(go.Scatter(
    x=df_iris.sepal_width, y=df_iris.sepal_length,
    mode='markers',
    marker_color=df_iris.sepal_width,
    text=df_iris.species, 
    marker=dict(showscale=True)
))
fig.update_traces(marker_line_width=2,marker_size=10)
````
